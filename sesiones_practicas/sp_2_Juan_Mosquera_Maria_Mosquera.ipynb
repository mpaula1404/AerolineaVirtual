### <span style="color:#2F749F;"><strong>Ejercicio 1: Cálculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528349` .  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:  
   - Media teórica: `(base % 10) + 2`  
   - Desviación estándar: `2`  
   - Tamaño de muestra: `300`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

image.png

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(528349)

media = (528349 % 10) + 2
desviacion = 2
muestra = np.random.normal(loc=media, scale=desviacion, size=300)
muestra


```

```python
# Medidas de tendencia central

mediana = np.median(muestra)
moda = stats.mode(muestra, keepdims=True)[0][0]

# Resultados
print("Medidas de tendencia central:")
print(f"Media:   {media:.2f} ")
print(f"Mediana: {mediana:.2f} ")
print(f"Moda:    {moda:.2f}")

plt.figure(figsize=(10, 5))
sns.histplot(muestra, bins=10, kde=False, color="#2F749F")

# Líneas de tendencia central
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='purple', linestyle=':', linewidth=2, label=f'Moda: {moda:.2f}')

plt.title("Distribución de ingresos por persona", fontsize=14)
plt.xlabel("Ingresos", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

image.png

```python
np.random.seed(528349)

media = (528349 % 10) + 2
desviacion = 2
sample_size = 300

ingresos = np.random.normal(loc=media, scale=desviacion, size=sample_size)
gastos = np.random.normal(loc=media, scale=desviacion, size=sample_size)

ingresos[ingresos < 0] = 0
gastos[gastos < 0] = 0

print("Ingresos:", ingresos)
print("\nGastos:", gastos)
```

**Coeficientes de correlación de Pearson y Spearman entre las matrices `ingresos` y `gastos`.**



```python
pearson_corr, _ = pearsonr(ingresos, gastos)
spearman_corr, _ = stats.spearmanr(ingresos, gastos)

print(f"Coeficiente de Correlación de Pearson: {pearson_corr:.2f}")
print(f"Coeficiente de Correlación de Spearman: {spearman_corr:.2f}")
```

#### Interpretación de los Coeficientes de Correlación

Los coeficientes de correlación nos ayudan a entender la relación entre las dos variables que hemos simulado, `ingresos` y `gastos`.

-   **Coeficiente de Correlación de Pearson (0.02)**:
    -   El coeficiente de Pearson mide la fuerza y dirección de una relación lineal entre dos variables. Su valor oscila entre -1 y +1.
    -   Un valor cercano a 0 como en este caso 0.02 indica que existe una relación lineal muy débil o nula entre los ingresos y los gastos. Esto significa que los cambios en los ingresos no están asociados de manera predecible con cambios lineales en los gastos, y viceversa, según esta muestra.
    -   El signo positivo sugiere una dirección positiva, aunque insignificante en este caso. Si el valor fuera más fuerte y positivo, implicaría que a medida que los ingresos aumentan, los gastos también tienden a aumentar. Un valor negativo fuerte indicaría lo contrario.

-   **Coeficiente de Correlación de Spearman (-0.00)**:
    -   El coeficiente de Spearman mide la fuerza y dirección de una relación monótona (no necesariamente lineal) entre dos variables. Se basa en el ranking de los datos en lugar de los valores brutos.
    -   Un valor cercano a 0 como en este caso  -0.00 también sugiere una relación monótona muy débil o nula entre los ingresos y los gastos. Esto indica que no hay una tendencia consistente en cómo el ranking de los ingresos se relaciona con el ranking de los gastos.
    -   El signo negativo (aunque extremadamente cercano a cero) no es significativo en este contexto. Si fuera un valor negativo fuerte, indicaría que a medida que el ranking de una variable aumenta, el de la otra tiende a disminuir.

Ambos coeficientes de correlación, Pearson y Spearman, son muy cercanos a cero. Esto sugiere que, en nuestra muestra simulada, **no hay una relación lineal ni monótona significativa** entre los ingresos y los gastos. Las dos variables parecen ser independientes entre sí, lo cual es esperable dado que fueron generadas de forma aleatoria e independiente con distribuciones normales.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=ingresos, y=gastos, alpha=0.6)
plt.title('Relación entre Ingresos y Gastos', fontsize=14, fontweight='bold')
plt.xlabel('Ingresos')
plt.ylabel('Gastos')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

image.png

```python
np.random.seed(528349)

educational_levels = np.random.choice(['Básico', 'Medio', 'Superior'], size=60, p=[0.3, 0.4, 0.3])
internet_access = np.random.choice(['Sí', 'No'], size=60, p=[0.7, 0.3])

data = pd.DataFrame({
    'Nivel educativo': educational_levels,
    'Acceso a internet': internet_access
})

print(data.head())
```

#### **Tabla de contigencia**:




```python
contingency_table = pd.crosstab(data['Nivel educativo'], data['Acceso a internet'])
print("Tabla de contingencia:")
print(contingency_table)
```

#### **Prueba de independencia de chi-cuadrado para determinar si existe una asociación significativa entre "Nivel educativo" y "Acceso a Internet"**:




```python
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-squared Test Results:")
print(f"Chi-squared Statistic: {chi2:.2f}")
print(f"P-value: {p:.2f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

#### Interpretación de los Resultados de la Prueba Chi-Cuadrado

La prueba de Chi-cuadrado de independencia se utiliza para determinar si existe una asociación estadísticamente significativa entre dos variables categóricas. Hemos analizado 'Nivel educativo' y 'Acceso a internet'.

-   **Estadístico Chi-cuadrado (0.21)**: Este valor mide la discrepancia entre las frecuencias observadas en la tabla de contingencia y las frecuencias esperadas si las dos variables fueran realmente independientes. Un valor más alto indica una mayor diferencia entre lo observado y lo esperado.
    -   En este caso, un valor de 0.21 es bastante bajo, lo que sugiere que las frecuencias observadas son muy similares a las que esperaríamos si no hubiera relación entre el nivel educativo y el acceso a internet.

-   **Valor p (0.90)**: El valor p es la probabilidad de observar un estadístico Chi-cuadrado tan extremo (o más extremo) como el calculado, asumiendo que la hipótesis nula (no hay asociación entre las variables) es verdadera.
    -   Un valor p de 0.90 es mucho mayor que los niveles de significancia comunes (como 0.05 o 0.01). Esto significa que **no tenemos evidencia suficiente para rechazar la hipótesis nula**. Es decir, no podemos concluir que exista una asociación estadísticamente significativa entre el nivel educativo y el acceso a internet en nuestra muestra.

-   **Grados de Libertad (2)**: Los grados de libertad (dof) se calculan como (número de filas - 1) * (número de columnas - 1) en la tabla de contingencia.

- En este caso, (3-1) * (2-1) = 2. Este valor es necesario para interpretar el estadístico Chi-cuadrado.


**Conclusión**:
Basándonos en los resultados de la prueba de Chi-cuadrado, con un valor p de 0.90, podemos concluir que **no existe una asociación estadísticamente significativa** entre el nivel educativo y el acceso a internet en esta muestra simulada. Las dos variables se consideran independientes.

# **Parte 3: Multivariado**

En este caso, el análisis multivariado no se desarrolla de manera más amplia debido a que el conjunto de datos únicamente contiene dos variables categóricas: nivel educativo y acceso a internet. El análisis multivariado generalmente requiere la inclusión de tres o más variables para poder evaluar relaciones simultáneas entre múltiples factores y comprender cómo interactúan entre sí. Al contar solo con dos variables, el análisis realizado corresponde principalmente a un **análisis bivariado**, el cual ya fue abordado mediante la tabla de contingencia, la visualización gráfica y la prueba de independencia Chi-cuadrado. Por esta razón, no se aplican técnicas multivariadas adicionales, ya que estas aportarían poca información adicional sin la incorporación de nuevas variables que permitan explorar relaciones más complejas dentro del conjunto de datos.


Para realizar el proceso de interpretación multivariado vamos añadir una nueva variable con datos simulados para lograr interpretación.

La tercera variables es la **edad**, con el fin de realizar un análisis de regresión logística multivariada. Este modelo permite evaluar la probabilidad de que una persona tenga acceso a internet en función de múltiples variables explicativas, en este caso el nivel educativo y la edad. Para ello, el nivel educativo fue transformado en variables dummy y la variable de acceso a internet se codificó de forma binaria. De esta manera, el análisis no solo considera la relación entre dos variables, sino que permite estudiar de manera simultánea el posible efecto de varios factores sobre la probabilidad de acceso a internet.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

np.random.seed(528349)

# Nivel educativo
educational_levels = np.random.choice(
    ['Básico', 'Medio', 'Superior'],
    size=60,
    p=[0.3, 0.4, 0.3]
)

# Edad simulada
edad = np.random.randint(18, 61, size=60)

# Asignar probabilidad de acceso a internet según nivel educativo y edad
prob_internet = []

for nivel, e in zip(educational_levels, edad):
    p = 0.5  # base

    if nivel == 'Medio':
        p += 0.15
    elif nivel == 'Superior':
        p += 0.25

    if e < 30:
        p += 0.10
    elif e > 50:
        p -= 0.10

    p = min(max(p, 0.05), 0.95)
    prob_internet.append(p)

internet_access = np.where(np.random.rand(60) < prob_internet, 'Sí', 'No')

# Crear DataFrame
data = pd.DataFrame({
    'Nivel educativo': educational_levels,
    'Edad': edad,
    'Acceso a internet': internet_access
})

# Variable dependiente binaria
data['internet_bin'] = data['Acceso a internet'].map({'Sí': 1, 'No': 0})

# Variables independientes
X = pd.get_dummies(data['Nivel educativo'], drop_first=True, dtype=int)
X['Edad'] = data['Edad']
X = sm.add_constant(X)

y = data['internet_bin']


modelo = sm.Logit(y, X).fit()

print(data.head())
print(modelo.summary())
```

Con base en los resultados del modelo de regresión logística, se analizó la probabilidad de que una persona tenga acceso a internet considerando como variables explicativas el nivel educativo y la edad. El modelo se estimó con 60 observaciones simuladas. Los coeficientes estimados indican que, en comparación con la categoría de referencia (nivel educativo básico), las personas con nivel medio y superior presentan coeficientes positivos, lo que sugiere una tendencia a una mayor probabilidad de tener acceso a internet. No obstante, los valores p asociados a estas variables (0.734 para nivel medio y 0.201 para nivel superior) son mayores al nivel de significancia común de 0.05, lo que indica que estas diferencias no son estadísticamente significativas en la muestra analizada. De manera similar, la variable edad presenta un coeficiente negativo (-0.0366), lo que sugiere que a mayor edad podría existir una ligera disminución en la probabilidad de acceso a internet; sin embargo, su valor p (0.129) tampoco resulta estadísticamente significativo. Además, el valor del Pseudo R-cuadrado (0.062) indica que el modelo explica una proporción relativamente baja de la variabilidad en el acceso a internet. En conjunto, estos resultados sugieren que, dentro de los datos simulados utilizados, ni el nivel educativo ni la edad muestran una influencia estadísticamente significativa sobre la probabilidad de tener acceso a internet, aunque se observan tendencias que podrían explorarse con muestras más grandes o con la inclusión de otras variables explicativas.
